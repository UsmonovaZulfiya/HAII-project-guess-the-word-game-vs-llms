```python
import transformers
import torch
from transformers import AutoTokenizer
```

```python
model = "meta-llama/Llama-3.1-8B-Instruct"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
generator = transformers.pipeline(
    "text-generation",
    model=model,
    dtype = torch.float16,
    device_map='auto',
)
```

```python
sequences = generator(
    'I want to travel somewhere, but don\'t have that much money. I am in Germany right now. Where can I travel with low cost?',
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    truncation = True,
    max_length = 400
)
```

```python
for seq in sequences:
    print(f"Result: {seq['generated_text']}")
```
